```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
sens = pd.read_csv("hmap_473.csv")
```

```python
xpow = np.logspace(-3.0, 0.0, 13)
```

```python
xlabels = np.linspace(-3.0, 0, 13)
```

```python
plt.figure(figsize = (10, 10))

plt.title("Sensitivity Measure $S(, \\gamma)$", fontsize=20, pad=10)

sb.heatmap(sens, cbar=True, square=True, cbar_kws={"shrink":0.8})


plt.axhline(y = 6.5, ls = 'dashed', c='black', lw=2)
#plt.axvline(x = 9.0, ymin = 0.385, ymax = 0.462, ls = '-', c='black', lw=2)
#plt.axvline(x = 10.0, ymin = 0.385, ymax = 0.462, ls = '-', c='black', lw=2)
#plt.axhline(y = 7.0, xmin=0.6935, xmax=0.7665, c='black', lw=2, ls='-')
#plt.axhline(y = 8.0, xmin=0.6935, xmax=0.7665, c='black', lw=2, ls='-')


plt.yticks(ticks = np.arange(0.5, 13.5, 1), labels = np.round(np.arange(0.3, -0.31, -0.05),3), fontsize=11)
#plt.yticks(ticks = np.arange(0.5, 5.5, 1), labels = [-0.2, -0.15, -0.1, -0.05, 0.0])

plt.xticks(ticks = np.arange(0.5, 13.5, 1), labels = xlabels, fontsize=11)


plt.xlabel("$ \\log_{10}\\frac{\\gamma}{\\nu}$", fontsize=18, labelpad=10)
plt.ylabel("Effective temperature, $$", fontsize=18, labelpad=10)

plt.savefig("fig3_v3.pdf", dpi = 300, bbox_inches='tight')
plt.show()
```
