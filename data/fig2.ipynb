```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
xarray = np.arange(1, 41)
```

```python
letlist = [['a', 'b'], ['c', 'd']]
```

```python
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(8, 6))

# Using seaborn's style
#plt.style.use('seaborn-v0_8-paper')
# With LaTex fonts
#plt.style.use('tex')
width = 345


for i in range(2):
    for j in range(2):
        hNum = pd.read_csv("hProfile_"+letlist[i][j]+".csv")
        yarray = hNum.mean()
        error = np.sqrt(hNum.var())

        axs[i][j].set_ylim(0, 1350)

        axs[i][j].plot(xarray, yarray, linewidth =2)
        axs[i][j].fill_between(xarray, yarray-error, yarray+error, alpha = 0.5)
        axs[i][j].tick_params(labelsize=16)

        #axs[i][j].set_xlabel('Site $i$', fontsize=20, labelpad=5)
        #axs[i][j].set_ylabel('Hb molecule number $x_i$', fontsize=20, labelpad=5)


axs[0][0].set_title("(a)", fontsize=28)
axs[0][0].text(20, 1150, "No communication", fontsize=26)
axs[0][0].text(26.75, 1000, "No feedback", fontsize=26)

axs[0][1].set_title("(b)", fontsize=28)
axs[0][1].text(20, 1150, "No communication", fontsize=26)
axs[0][1].text(22.5, 1000, "Strong feedback", fontsize=26)

axs[1][0].set_title("(c)", fontsize=28)
axs[1][0].text(15.75, 1150, "Strong communication", fontsize=26)
axs[1][0].text(26.75, 1000, "No feedback", fontsize=26)

axs[1][1].set_title("(d)", fontsize=28)
axs[1][1].text(32, 1150, "Optimal", fontsize=26)

fig.text(1, -0.02, 'Site $i$', fontsize=28)
fig.text(-0.02, 0.5, 'Hb molecule number $x_i$', fontsize=28, rotation='vertical')

plt.subplots_adjust(left  = 0.125, right = 2, bottom = 0.1, top = 1.5, wspace = 0.15, hspace = 0.25)

plt.savefig("fig2_v2.pdf", dpi=300, bbox_inches ='tight')
plt.show()
```
