```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# Theta vs. A (amplitude of the tanh initial condition)
theta_arr = np.linspace(0.3, -0.3, 13)
A_arr = np.arange(0, 750, 50)

theta_vs_A = np.genfromtxt('hmap_theta_vs_A.csv', delimiter=',')

thetaStar_A = np.zeros(len(A_arr))

for j in range(len(A_arr)):
    thetaStar_A[j] = theta_arr[np.where(theta_vs_A[j] == theta_vs_A[j].max())[0][0]]
```

```python
plt.figure(figsize=(10,6))
plt.plot(A_arr, thetaStar_A, lw=2)
plt.ylim((-0.35, 0.1))
plt.xticks(A_arr)
plt.xlabel('$A$', fontsize=16)
plt.ylabel('$\\theta^*$', fontsize=16)
plt.title('$\\theta^*$ vs. Amplitude $A$', fontsize=16)
plt.savefig('imgs/thetaStar_vs_A.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
theta_arr = np.linspace(0.3, -0.3, 13)
H = 2.0
xc_arr = np.around((np.arange(164, 1641, 164, dtype=int))*(((H-1)/(H+1))**(1/H)))
```

```python
theta_vs_xc = np.genfromtxt('hmap_theta_vs_xc.csv', delimiter=',')
thetaStar_xc = np.zeros(len(xc_arr))

for j in range(len(xc_arr)):
    thetaStar_xc[j] = theta_arr[np.where(theta_vs_xc[:,j] == theta_vs_xc[:,j].max())[0][0]]
```

```python
plt.figure(figsize=(10,6))
plt.plot(xc_arr, thetaStar_xc, lw=2)
plt.ylim((-0.3, 0.2))
plt.xticks(xc_arr)
plt.xlabel('$x_c$', fontsize=16)
plt.ylabel('$\\theta^*$', fontsize=16)
plt.title('$\\theta^*$ vs. $x_c$', fontsize=16)
plt.savefig('imgs/thetaStar_vs_xc.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
